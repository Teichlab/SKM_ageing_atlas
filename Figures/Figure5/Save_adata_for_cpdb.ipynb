## Separate cellphonedb input into young and old

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scanpy as sc
```

```python
adata = sc.read("cpdb/SKM_myofiber2immcells/SKM_myofiber2immcells_adata_2021-07-04.h5ad")
```

```python
adata.obs[['Age', 'Age_bin', 'DonorID', 'batch']].drop_duplicates().sort_values(['Age'])
```

```python
adata.obs['Age_bin'] = adata.obs['Age'].apply(lambda x: 'young' if x < 40 else 'old')
```

```python
adata.obs.groupby(['Age', 'Age_bin']).count()
```

```python
adata_young = adata[adata.obs['Age_bin'] == 'young',:].copy()
```

```python
adata_old = adata[adata.obs['Age_bin'] == 'old',:].copy()
```

```python
adata.shape
adata_young.shape
adata_old.shape
```

```python
adata_young.obs['annotation_subcluster'].value_counts()
```

```python
adata_young.obs['annotation_subcluster'].unique()
```

```python
adata_old.obs['annotation_subcluster'].unique()
```

```python
[i for i in adata_old.obs['annotation_subcluster'].unique() if i not in adata_young.obs['annotation_subcluster'].unique()]
```

```python
adata_young.obs['cpdb_ctype*age'] = adata_young.obs['annotation_subcluster'].astype(str) + '-'\
+ adata_young.obs['Age_bin'].astype(str)
```

```python
adata_old.obs['cpdb_ctype*age'] = adata_old.obs['annotation_subcluster'].astype(str) + '-'\
+ adata_old.obs['Age_bin'].astype(str)
```

```python
adata_young.obs[['annotation_subcluster', 'cpdb_ctype*age']]
```

```python
adata_old.obs[['annotation_subcluster', 'cpdb_ctype*age']]
```

```python
adata_young.write("cpdb/SKM_myofiber2immcells_young/SKM_myofiber2immcells_adata_young_"+today +".h5ad")
```

```python
df_meta_young = pd.DataFrame(data={'Cell': list(adata_young.obs_names),
                       'cell_type': list(adata_young.obs['cpdb_ctype*age'])})
df_meta_young.set_index('Cell',inplace=True)

df_meta_young['cell_type'].unique()

df_meta_young.to_csv("cpdb/SKM_myofiber2immcells_young/SKM_myofiber2immcells_metadata_young_" + today + ".tsv", sep='\t')
```

```python
adata_old.write("cpdb/SKM_myofiber2immcells_old/SKM_myofiber2immcells_adata_old_"+today +".h5ad")
```

```python
df_meta_old = pd.DataFrame(data={'Cell': list(adata_old.obs_names),
                       'cell_type': list(adata_old.obs['cpdb_ctype*age'])})
df_meta_old.set_index('Cell',inplace=True)
df_meta_old['cell_type'].unique()
df_meta_old.to_csv("cpdb/SKM_myofiber2immcells_old/SKM_myofiber2immcells_metadata_old_" + today + ".tsv", sep='\t')
```
